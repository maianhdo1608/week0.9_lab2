```python
import pandas as pd

# Sử dụng các tham số bổ sung để bỏ qua dòng lỗi và tự động nhận diện cấu trúc
try:
    data = pd.read_csv('word,meaning.csv',
                       on_bad_lines='skip',
                       engine='python',
                       encoding='utf-8')

    print("Dữ liệu đã sẵn sàng (đã tự động bỏ qua các dòng bị lỗi cấu trúc)!")
    print(f"Tổng số từ vựng đọc được: {len(data)} từ")
    print(data.head())
except Exception as e:
    print(f"Vẫn còn lỗi: {e}")
```

```python
import random

# 1. Chuyển đổi dữ liệu từ bảng (DataFrame) sang danh sách (List) để dễ xáo trộn
danh_sach_hoc = data.values.tolist()

# 2. Trộn ngẫu nhiên các từ vựng để bạn không học vẹt theo thứ tự
random.shuffle(danh_sach_hoc)

print("--- CHÀO MỪNG BẠN ĐẾN VỚI CHATBOT FLASHCARD ---")
print("Hướng dẫn: Nhập nghĩa tiếng Việt hoặc gõ '종료' để kết thúc.")

# 3. Vòng lặp: Máy sẽ hỏi lần lượt từng từ
for dong in danh_sach_hoc:
    tu_han = dong[0]      # Cột 1 là tiếng Hàn
    nghia_dung = dong[1]  # Cột 2 là nghĩa tiếng Việt

    # Hiển thị câu hỏi và chờ bạn nhập câu trả lời
    cau_tra_loi = input(f"\nTừ này nghĩa là gì: {tu_han}? ")

    # Kiểm tra nếu bạn muốn dừng lại
    if cau_tra_loi == '종료':
        print("\nHẹn gặp lại bạn lần sau. Chăm chỉ nhé!")
        break

    # So sánh câu trả lời của bạn với đáp án đúng
    # .strip() giúp loại bỏ khoảng trắng thừa nếu bạn vô tình gõ dư
    if cau_tra_loi.strip() == str(nghia_dung).strip():
        print("✅ Chính xác! Giỏi quá.")
    else:
        print(f"❌ Sai rồi. Nghĩa đúng phải là: {nghia_dung}")

print("\n--- BẠN ĐÃ HOÀN THÀNH DANH SÁCH TỪ VỰNG! ---")
```
